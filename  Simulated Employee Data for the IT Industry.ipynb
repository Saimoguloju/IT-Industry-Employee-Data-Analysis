```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender', palette='viridis')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig("Distribution of Gender.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Experience (Years)'], kde=True, color='teal')
plt.title('Distribution of Experience (Years)')
plt.xlabel('Experience (Years)')
plt.ylabel('Frequency')
plt.savefig("Distribution of Experience (Years).png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True, color='orange')
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.savefig("Distribution of Salary.png")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Position', palette='magma', order=df['Position'].value_counts().index)
plt.title('Count of Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig("Count of Positions.png")
plt.show()
```

```python
avg_salary_by_position = df.groupby('Position')['Salary'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=avg_salary_by_position, x='Position', y='Salary', palette='coolwarm')
plt.title('Average Salary by Position')
plt.xlabel('Position')
plt.ylabel('Average Salary')
plt.savefig("Average Salary by Position.png")
plt.xticks(rotation=90)
plt.show()
```

```python
avg_salary_by_gender = df.groupby('Gender')['Salary'].mean().reset_index()
plt.figure(figsize=(6, 4))
sns.barplot(data=avg_salary_by_gender, x='Gender', y='Salary', palette='pastel')
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.savefig("Average Salary by Gender.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Experience (Years)', y='Salary', hue='Gender', palette='Set1', alpha=0.7)
plt.title('Experience vs Salary')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.legend(title='Gender')
plt.savefig("Experience vs Salary.png")
plt.show()
```

```python
# Compute the correlation matrix
corr = df[['Experience (Years)', 'Salary']].corr()
# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.savefig("Correlation Matrix.png")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Position', hue='Gender', palette='dark')
plt.title('Gender Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.savefig("Gender Distribution by Position.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Experience (Years)', hue='Gender', multiple='stack', palette='tab10')
plt.title('Experience Distribution by Gender')
plt.xlabel('Experience (Years)')
plt.ylabel('Frequency')
plt.savefig("Experience Distribution by Gender.png")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Position', y='Salary', palette='crest')
plt.title('Salary Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.savefig("Salary Distribution by Position.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Experience (Years)', y='Salary', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Experience vs Salary with Regression Line')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.savefig("Experience vs Salary with Regression Line.png")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lmplot(data=df, x='Experience (Years)', y='Salary', hue='Gender', palette='Set1', aspect=1.5, height=6)
plt.title('Experience vs Salary by Gender with Regression Lines')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.savefig("Experience vs Salary by Gender with Regression Lines.png")
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Position', y='Salary', hue='Gender', palette='muted')
plt.title('Salary Distribution by Gender and Position')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.savefig("Salary Distribution by Gender and Position.png")
plt.show()
```

```python
g = sns.FacetGrid(df, col='Position', col_wrap=3, height=8, aspect=1.5)
g.map_dataframe(sns.scatterplot, x='Experience (Years)', y='Salary', hue='Gender', palette='tab10', alpha=0.7)
g.add_legend()
g.set_axis_labels('Experience (Years)', 'Salary')
g.set_titles(col_template="{col_name}")
plt.savefig("Experience (Years) & Salary.png")
plt.show()
```
